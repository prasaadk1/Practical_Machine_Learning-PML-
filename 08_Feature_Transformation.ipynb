### i) Log Transformation
- Converting each value to its logarithmic value
- Before applying log we have to check the minimum value if it is zero then we have to convert it into non zero
- If it is zero then add 00001 in each value

```python
np.log(df["avg_glucose_level"])
```

```python
df["avg_log"]= np.log(df["avg_glucose_level"])
```

```python
df["avg_log"].skew()
```

### ii) Root Transformation

```python
df["root_agl"]= df["avg_glucose_level"]**(1/12)
df["root_agl"].skew()
```
