# **Measures of central x (1st Business Moment)**

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame({"X":[1,1,2,5,4,3],"Y":[1,1,2,3,4,4],"Z":[1,1,2,2,3,3],"I":[5,1,3,4,2,np.nan]})
```

### **1) Mean**   

**- Applied on Continuous Data**

- **Polulation Mean = (µ) = ∑ X / N**
- **Sample Mean = (µ) = ∑ X / n**

```python
df
```

```python
# Option 1
# We are appliying mean for "X " Column\
df["X"].sum()/len(df["X"])
```

```python
# Option 2
df["X"].mean()
```

### **2) Median**   

- **Also Applied on Continuous Data**
- **Median refers to the data value that is positioned in the middle of ordered dataset**
- Median refers to centre if odd number of data points
- Median refers to average of centre 2 values if even number of data points

```python
df["X"].median()
```

### **3) Mode**   

**- It is applied on discrete variable**

**- It is a most repeated or we can say frequent value**

1) **Unimodal Data** - ( if the data have only 1 Mode Value)
2) **Bimodal Data** - (if the data have 2 Mode Value)
3) **Multimodal Data** - (If the data have greater than 2 Mode value)

```python
df["X"].mode()
```

```python
df["Y"].mode()
```

```python
df["Y"].mode()
```

```python
df["I"].mode()
```
