#### **1) Functions for Creating 2 Dimentional (2D) arrays**

**1) Zeros**


```python
np.zeros((3,3))
```

We can also Perform Operations on this 

```python
np.zeros((3,3),dtype = int) +1
```

Like that we are abe to set Datatype and along with it we can perform mathematical operation

**2) Ones**


```python
np.ones((3,3))
```

We can also Perform Operations on this same as zeros  

```python
np.ones((3,3),dtype = int) +1
```

**3) Identity Matrix**
- An identity matrix is a square matrix with ones on the diagonal and zeros elsewhere, acting as the multiplicative identity in matrix operations.


```python
np.eye(3,3)
```

**4) Diagonal Matrix**

```python
np.diag([1,2,3])
```

```python
a = np.array([[1,2,3],[3,4,5]])
np.diag(a)
```

#### **2) Accessing elements / Extracting values in  2 Dimentional (2D) arrays**

```python
a = np.array([[10,20,30],[30,40,50],[20,80,90]])
a
```

**1) To extract a single value from 2D array** 

- syntax: arrayobj[row index number , column index number]

```python
a[1,2]
```

**2) To extract a single row**
- syntax: arrayobj[row index number]

```python
a[2]
```

**3) To extract multiple rows**
- syntax: arrayobj[start:end]

```python
a[1:3]
```

#### **4) Operations in  2 Dimentional (2D) arrays**

**1) Mathematical Operations**

```python
a = np.array([[1,2],[3,4],[5,6]])
b = np.array([[40,4],[14,24],[34,54]])
```

- Addition

```python
print(a+b)
```

- Multiplication

```python
print(a*b)
```

- Matrix Multiplication

```python
a = np.array([[1,2,1],[3,4,5],[5,6,5]])
b = np.array([[4,4,1],[4,4,2],[4,4,3]])
np.matmul(a,b)
```

#### **5) Join Multiple  (2D) arrays**

```python
a = np.array([[1,2,1],[3,4,5],[5,6,5]])
b = np.array([[4,4,1],[4,4,2],[4,4,3]])
```

```python
np.concatenate((a,b),axis = 0)
```

**- Vertical stack**

```python
a = np.array([[1,2,1],[3,4,5],[5,6,5]])
b = np.array([[4,4,1],[4,4,2],[4,4,3]])
np.vstack((a,b))
```

**- Horizontal stack**

```python
a = np.array([[1,2,1],[3,4,5],[5,6,5]])
b = np.array([[4,4,1],[4,4,2],[4,4,3]])
np.hstack((a,b))
```
