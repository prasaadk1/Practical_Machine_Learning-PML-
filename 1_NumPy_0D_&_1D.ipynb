#### **2.1)  Functions for Creating 1 Dimentional (1D) arrays**

**1) Arange Function**
   - Arange function is a array valued version of the build in Python range function.
   - **Syntax** - np.arange(start,end,step)

```python
a = np.arange(1,10,2)
```

- Basically it gives values in range of 1 to 10 with step of 2 see the following output.

```python
a
```

**2) Linspace**
   - It returns evenly spaced numbers over a specified interval.
   - **Syntax** - np.linspace(start,end,num_of_samples)

```python
np.linspace(0,5,6)
```

- In the above example we are getting 6 samples evenly spaced

**3) Zeros**
   - It generates arrays of zeros.
   - **Syntax** - np.zeros(shape)

*Option 1*

```python
np.zeros((4,))
# By default data type is Float
```

*Option 2*

```python
np.zeros((4,),dtype = int)
```

**4) Ones**
   - It generates arrays of ones.
   - **Syntax** - np.ones(shape)

*Option 1*

```python
np.ones((4,))
```

*Option 2*

```python
np.ones((4,),dtype = int)
```

**4) random.randint**
   - Return random intergers from start to end 
   - **Syntax** - np.random.randint(start,end,num_of_samples)

```python
np.random.randint(1,100,10)
```

#### **2.2) Accessing elements / Extracting values in  1 Dimentional (1D) arrays**

```python
a = np.array([1,2,4,5,6,100,120 ,34,56,78])
a 
```

**1) Accessing a single value through indexing**

Suppose we want to extract the element at index 2 we use following input

```python
a[1]
```

**2) Accessing multiple value through slicing**

```python
a[0:5:2]
```

**3) Accessing multiple value through indexing**

syntax: arrayobj[[list of index numbers sep by comma]]

```python
a[[1,2,5,6]]
```

**4) Accessing multiple value Based on condition (Mask indexing or Fancy Indexing**

Syntax: arrayobj [condition]

```python
a[a%2]
```

```python
a[a>20]
```

```python
a[a==100]
```

#### **2.3) Modify the values in  1 Dimentional (1D) arrays**

1) Add single value in array
2) Add multiple value in array
3) Replace a single value in array
4) Replace multiple values with a single in array
5) Replace multiple values with multiple values  in array
6) Delete multiple values in an array
7) Copy array
8) Sorting of array

**1) Add single value in array**

Syntax: np.append(arrayobj,value)

```python
a = np.array([45,65,23,34,543,324])
a1 = np.append(a,110)
a1
```

**2) Add multiple value in array**

Syntax: np.append(arrayobj,[values sep by comma])

```python
a2 = np.append(a,[11,22,44,55,77,88])
a2
```

**3) Replace a single value in array**

```python
a = np.array([45,65,23,34,543,324])
a[1] = 50000
a
```

**4) Replace multiple values with a single value in array**

```python
arrayobj[index numbers sep by comma] = value
```

```python
a[[2,1]] = 0
a
```

**5) Replace multiple values with multiple values in array**

```python
a5 = np.array([45,65,23,34,543,324])
a5 [[1,2]] = [10,20]
a5
```

**6) Delete multiple values in an array**

Syntax : np.delete(array,index number)

```python
a = np.array([1,2,3,4])
np.delete(a,1)
```

**7) Copy array**

Shallow Copy

```python
a= np.array([23,34,65,67])
b = a.copy() 
```

```python
print(a,id(a))
print(b,id(b))
```

Deep Copy

```python
c = a
print(c,id(c))
```

**7) Sorting an array**

```python
a= np.array([23,34,65,67])
b = np.sort(a)
b
```
