#### **2.5) Select/ Extract the required data**

```python
df.loc[0:4]
```

**2.5.3) Extract a multiple row without slicing**

```python
df.loc[[0,1,4]]
```

**2.5.4) Extract a single column**

```python
df.loc[1:,"Age"]
```

**2.5.5) Extract a multiple column slicing based**

```python
df.loc[:,["Age","Gender"]]
```

- df.loc[[multiple row sep by comaa],[multiple column sep by comma]]

#### **2.7) Modify the DataFrames**

```python
# Example
d = {"Age":[22,23,25,24,26],
    "Salary": [170000,150000,250000,350000,460000],
    "Gender": ["M","F","M","M","F"]}
df = pd.DataFrame(d)
df
```

**2.7.1) Drop a single row**

```python
df = df.drop(index = [2])
df
```

```python
df
```

```python
df.drop(index = [0],inplace = True)
df
```

- In the given line of code, the inplace=True argument means that the changes made to the DataFrame (dropping the specified row) will directly modify the original DataFrame without creating a new one.
  
- If you do not specify inplace=True in the drop() method (or other similar pandas operations), the default behavior is inplace=False. This means the method will not modify the original DataFrame but instead return a new DataFrame with the specified changes.

```python
# Example
d = {"Age":[22,23,25,24,26],
    "Salary": [170000,150000,250000,350000,460000],
    "Gender": ["M","F","M","M","F"]}
df = pd.DataFrame(d)
df
```

**2.7.2) Drop Multiple rows**

```python
df1 = df.drop(index = [0,1,2])
df1
```

**2.7.3) Drop Single Column**

```python
df
```

```python
df2 = df.drop(columns = ['Age'])
df2
```

**2.7.4) Drop Multiple Columns**

```python
df3 = df.drop(columns=['Age','Gender'])
df3
```

#### **2.9)Combining Dataframes**

```python
df1 = pd.DataFrame({"City":["Mumbai","Delhi","Banglore","Hyderabad"],"Temperature":[34,20,40,37]})
df1
```

```python
df2 = pd.DataFrame({"City":["Delhi","Mumbai","Banglore","Hyderabad","Chennai"],"Humidity":[68,56,75,70,45]})
df2
```

**2.9.1) Concatenation of Dataframes**

- axis=0: This means the operation is performed along the rows (vertical concatenation.
- axis=1: This means the operation is performed along the columns (horizontal concatenation).

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis = 0)
```

- When ignore_index=True, the index is reset to a continuous sequence (0, 1, 2, 3).

```python
pd.concat([df1,df2],axis = 0,ignore_index = True)
```

**2.9.2) Merging of Dataframes**

```python
df1
```

```python
df2
```

**1)Left Join**

```python
pd.merge(df1,df2,on ="City",how= "left")
```

**2)Right Join**

```python
pd.merge(df1,df2,on ="City",how= "right")
```

**3)Inner Join**

```python
pd.merge(df1,df2,on ="City",how= "inner")
```

**4)Outer Join**

```python
pd.merge(df1,df2,on ="City",how= "outer")
```
