### Outlier Detection

- This Data is used fro supervised Learning
- Target Column = 'Attrition'
- All other columns are features or input comlumns(X)

```python
df_ohe.columns
```

```python
X = df_ohe.drop(['Attrition_Yes','Attrition_No'],axis = 1)
Y = df_ohe['Attrition_Yes']
```

```python
X.shape , Y.shape
```

## Split the data in training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,
                                    test_size=0.3,
                                    random_state=7)
```

```python
X_train.shape, X_test.shape, Y_train.shape,Y_test.shape
```

## Standardize the continuous columns

- standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

sc.fit(X_train,Y_train)
```

```python

X_train_sc = X_train.copy()
X_test_sc = X_test.copy()
X_train_sc.iloc[:,:] = sc.transform(X_train.iloc[:,:])
X_test_sc.iloc[:,:] = sc.transform(X_test.iloc[:,:])
```
