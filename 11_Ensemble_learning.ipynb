```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/prasad/Prasad Folder/DBDA/PML/cleaned_76hd.csv')
df
```

```python
X = df.drop("V58",axis=1)
Y = df['V58']
```

```python
X.shape, Y.shape
```

### **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,
                                    test_size=0.3,
                                    random_state=7)
```

```python
X_train.shape, X_test.shape, Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression(random_state = 7)
lr.fit(X_train, Y_train)
Y_pred_lr = lr.predict(X_test)
```

```python
gb = GaussianNB()
gb.fit(X_train, Y_train)
Y_pred_gb = gb.predict(X_test)
```

```python
sv = SVC(random_state = 7)
sv.fit(X_train, Y_train)
Y_pred_sv = sv.predict(X_test)
```

```python
kn = KNeighborsClassifier()
kn.fit(X_train, Y_train)
Y_pred_kn = kn.predict(X_test)
```

```python
dt = DecisionTreeClassifier(random_state=7)
dt.fit(X_train, Y_train)
Y_pred_dt = dt.predict(X_test)
```

### Apply Ensemble Learning Models 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_samples = 0.8, oob_score = True,
                       random_state=7)
```

```python
rf.fit(X_train,Y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print(classification_report(Y_test,y_pred ))
```

# XGBoost

```python
import xgboost as xg
```

```python
xgc = xg.XGBClassifier(random_state=7)
```

```python
xgc.fit(X_train,Y_train)
```

```python
y_pred = xgc.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(Y_test,y_pred, average="micro" )
```

```python
print(classification_report(Y_test,y_pred ))
```

```python
xgc = xg.XGBClassifier(
                       learning_rate=0.1,
                       n_estimators=100,
                       max_depth=5,
                       subsample=0.8,
                       colsample_bytree=0.8,
                       random_state=7)
xgc.fit(X_train,Y_train)
```

```python

y_pred = xgc.predict(X_test)
print(classification_report(Y_test,y_pred ))
```
