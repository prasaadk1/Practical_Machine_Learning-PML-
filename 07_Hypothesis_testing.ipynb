# Z test 

**case study 1:-** The length of 25 samples of a fabric are taken at random. Historic Mean and standard deviation are 150 cm and 5.5 cm respectively. Company manager wants to test if the current mean is equal to the the historic mean. Assume a to be 0.05

**Step1:**- Formation of Ho, Ha

Ho : Mean = 150 

Ha : Mean != 150

**Step2:**- Select level of significance

alpha = 0.05

**Step3:**- Check for Normality


```python
df1 = pd.read_excel("Fabric_data.xlsx")
df1.head()
```

```python
df1["Fabric_length"].skew()
```

**Step4:-** Select the statistical test and calculate p value.

- 1 Sample Z test 2 tail rejection.

```python
from statsmodels.stats.weightstats import ztest
```

```python
ztest = ztest(df1["Fabric_length"],
value=150,
alternative = 'two-sided',
ddof=0)
```

```python
ztest
```

(test statstic , P value)

**Step 5-** Based on P value,Accept or Reject

# T test
**Case Study 2:** The mean diameter of the bolt manufactured should be 10 mm to be able fit into the nut 20 samples are taken at random from production line bya quality inspector conduct a test to check with 90% confidence that the mean is not different from the specification value.

**1) Formulation of Ho and Ha**

- Ho Mean = 10 (no action)
- Ha Mean != 10 (Action)

**2) Select level of significance**
- alpha = 0.1

**3) Check for Normalilty**

```python
df2 = pd.read_excel("Bolt diameter.xlsx")
```

```python
df2.head()
```

```python
df2["Diameter"].skew()
```

**4) select the statistical test and calculate the p value**
- 1 sample test T test with two tail Rejection.

```python
from scipy import stats
ttest = stats.ttest_1samp(df2["Diameter"],popmean = 10,alternative = "two-sided")
ttest
```

# Anova Test

**Case Study 3** A financial analyst at a financial institute wants to evaluate a recent credit card promotion after this promotion 500 cardholders were randomly selected Half received and ad promoting a full wavier of interest rate on purchases made over the next three months and half received a standard chrismas ad did the ad promoting full interest rate wavier increase purchase?

**1) Formulation of Ho,Ha**

Ho : (No difference): Avg of purchases made by FW <= Avg Purchases made sc => default / current/ no action

Ha : Avg of Purchases made by F/W > Avg purchases made by sc => take action.

**2) Select Level of Significance**
- alpha = 0.05

**3) check for Normality**

```python
df3 = pd.read_excel("Promotion.xlsx")
df3.head()
```

```python
df3["InterestRateWaiver"].skew()
```

```python
df3["StandardPromotion"].skew()
```

**4) Select the statistical test and calculate the p value**
- Annova Test (F- Test)

```python
anova = stats.f_oneway(df3["InterestRateWaiver"],df3["StandardPromotion"])
anova
```

- Based on p value accept or reject

# Proportion Test

**Case Study 4-** The people carry out a poll to find the acceptability of new football coach. It was decided that if the support rate for the coach for the entire population was truly less than 90% the coach would be fired 2k people participated and 496 people supported the new coach. conduct a test to check if the New coach should be fired with 95% level of confidence

**1) Formulation of Ho,Ha**

Ho : coach not to be fired (No action) No difference

H1 : coach to fired (action)

**2) Select level of Significance**
- alpha = 0.05

**4) select the statistical test and calculate the p value**

-1 Proportion test

```python
binom_test = stats.binomtest(496,2000,0.95)
binom_test
```

- Based on p value accept or reject

The sales executives felt that adults (>40 yrs) won't buy, children will & hence requested sales manager not to launch the program. Analyze the data & determine whether there is evidence at 5% significance level to support the hypothesis

**1. Formulation of H0,H1**

• HO -> Proportions of Adults = Proportions of Children

• H1 -> Proportions of Adults != Proportions of Children


**2. Select Level of significance**
    
• alpha = 0.05


**4. Select the Statistical test & Calculate the p value**
• Chi-Square Test


```python
df5=pd.read_excel("JohnyTalkers.xlsx")
df5. head()
```

```python
df5.info()
```

```python
count = pd.crosstab(df5["Person"], df5["Drinks"],margins=True)
count
```

```python
chisquare_test = stats. chi2_contingency (count)
chisquare_test
```

**5. Based on p value, Accept or Reject HO**

• P < alpha
