# **Data Preprocessing**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/prasad/Prasad Folder/DBDA/Machine Learning/BankChurners.csv',na_values= ['unknown','unknown'])
```

```python
df
```

```python
df.info()
```

### 1. Dropping the Output column 

```python
X = df.drop(['Attrition_Flag'],axis = 1)
```

```python
y = df['Attrition_Flag']
```

### 2. Check the null values

```python
X.isnull().sum()
```

```python
X.nunique()
```

### 3. Remove Duplicates  

```python
X.drop_duplicates(inplace = True)
```

```python
X.shape
```

```python
X.dtypes
```

#### 4.Apply one hot encoding basically Data Encoding

```python
X_ohe = pd.get_dummies(X)
```

```python
X_ohe.shape
```

```python
X_ohe.columns
```

```python
X_ohe.head()
```

```python
X_ohe_std = (X_ohe - X_ohe.mean())/X_ohe.std()
```

```python
X_ohe_std.mean()
```

```python
y.unique()
```

```python
y = y.replace({'Existing Customer':0,
           'Attrited Customer':1})
```

```python
y.unique()
```

### **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_ohe_std
                                                 ,y,
                                    test_size=0.3,
                                    random_state=7)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

### **Fine Tuning using Grid Search**

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'C' : [5,10,20,40,50,60],
 'kernel':['rbf'],
'gamma':[10,52,100]}

gs = GridSearchCV(SVC( random_state=7),param_grid = param_grid, cv =2 , verbose = 2)
```

```python
gs.fit(X_train,Y_train)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
# C 10 gamma 0.1  --> 92.8
# C 20 gamma 0.01 --> 93
```

```python
# Create model with best params

svc = SVC(C=20,kernel='rbf',gamma=0.01,random_state=7)
svc.fit(X_train,Y_train)
y_predict = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predict))
```

# **Fine Tune Random Forest**

- n_estimators [100,500,1000]
- max_features ['sqrt','log']
- max_depth [ 5,10,15]
- oob_score ['True']
- max_samples [0.8]

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state = 7)
```

```python
param_grid = {
    'n_estimators': [100, 500,1000],
    'max_depth': [5,10,15],
    'oob_score' : [True],
    'max_samples':[0.8],
    'max_features': ['sqrt', 'log2']}
```

```python
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=2)
```

```python
grid_search.fit(X_train, Y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
grid_search.best_score_
```
