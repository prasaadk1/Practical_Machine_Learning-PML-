### **4) Filtering**
    - selecting/extracting the data based on conditions. 

**syntax:** df[condition]

```python
# Here we are Extracting the data in with the tip is greater than 7
df[df["tip"]>=7]
```

```python
# Here we are Extracting the data in which only male are there
df[df["gender"]== "Male"]
```

#### **4.1) Filtering using multiple conditions**

**Syntax:** df[(condition1) or'|' and '&' (condition2)]

```python
# Here we have extracted total bill is greater than 40 and the gender is female
df[(df["total_bill"]>40) & (df["gender"]=="Female" )]
```

```python
# Here we have extracted total bill is greater than 40 and the gender is Male
df[(df["total_bill"]>40) & (df["gender"]=="Male")]
```

```python
# Here we have extracted tip is greater than or equal to 7 or members should be equal to 6
df[(df['tip']>=7) | (df['no_of_members']==6)]
```

```python
# Here we have extracted total bill is greater than or equal to  40 or tip is greater than or equal to 7  or the gender is Male
df[(df['total_bill']>=40) | (df['tip']>=7) & (df['gender']=="Female" )]
```

```python
# Here we have extracted day is sunday and gender is male
df[(df["day"]=="Sun") & (df["gender"]=="Male" )]
```

```python
# Here we have extracted day is sunday and gender is Female

df[ (df["day"]=="Sun") & (df["gender"]=="Female")]
```

### **6) Discretization**
    - Converting a numerical data to non numerical data

-So the Following code will create a new column named bill category and assigns the category of their bill means in bins we have declared that 0 to 10 is low and 10 to 30 is moderate and 30 to 50 is good accordingly it assigns the category and integrate the catgeory into the new column


```python
df['bill_category'] = pd.cut(df['total_bill'],
                bins = [0,10,30,51],
                labels= ["low", "moderate" , "good"])
df
```

```python
# This function will show the count of each category that we have recently created
df["bill_category"].value_counts()
```
